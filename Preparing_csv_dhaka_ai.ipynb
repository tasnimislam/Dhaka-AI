```
%cd "/content/drive/My Drive"
#!mkdir coco_label
destination = "/content/drive/My Drive/coco_label"
IMAGE_WIDTH = 512
IMAGE_HIEGHT = 512
def convert_to_coco(xml_directory):
  print(xml_directory)
  content = []
  with open(xml_directory, "r") as file:
    # Read each line in the file, readlines() returns a list of lines
    content = file.readlines()
    # Combine the lines in the list into a string
    content = "".join(content)
    bs_content = bs(content, "lxml")

    filename = bs_content.filename
    source = bs_content.source
    size = bs_content.size
    height = size.height
    width = size.width
    

    width=int(width.get_text())
    height=int(height.get_text())

    print(height)
    print(width)

    segmented = bs_content.segmented
    txt_name=xml_directory.split("/")
    txt_name=txt_name[-1]
    txt_name=txt_name.split(".")
    txt_name=txt_name[0] + ".txt"
    file= open(txt_name,"w+")
    for obj in bs_content.find_all('object'):
      name = obj.find("name")

      if class_no_dao(name.get_text()) == 1000:
        continue

      xmin=obj.find("xmin")
      xmin=xmin.get_text()
      xmin=int(xmin)
      xmin=xmin/width

      ymin=obj.find("ymin")
      ymin=ymin.get_text()
      ymin=int(ymin)/height
      #ymin=int(int(ymin)*IMAGE_HIEGHT/height)

      xmax=obj.find("xmax")
      xmax=xmax.get_text()
      xmax=int(xmax)/width
      #xmax=int(int(xmax)*IMAGE_WIDTH/width)

      ymax=obj.find("ymax")
      ymax=ymax.get_text()
      ymax=int(ymax)/height
      #ymax=int(int(ymax)*IMAGE_HIEGHT/height)

      x_center = (xmax+xmin)/2
      y_center = (ymax+ymin)/2

      w=(xmax-xmin)/2
      h=(ymax-ymin)/2

      file.write(str(class_no_dao(name.get_text())) + " " + str(x_center) + " " + str(y_center) + " " + str(w) + " " +str(h))
      file.write("\n")
  file.close()
  shutil.move(txt_name, destination)
```
