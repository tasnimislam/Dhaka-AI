```
import pandas as pd
df = pd.DataFrame(columns = ["image_name", "x1", "y1", "x2", "y2", "class_name"])
listi = []
```

```
import cv2
from google.colab.patches import cv2_imshow
image = cv2.imread("/content/drive/MyDrive/df_fold_1/images/train_image/02.jpg")
image_name = "/content/drive/MyDrive/df_fold_1/images/train_image/02.jpg"
image_shape = image.shape
print(image_shape)
file = open("/content/drive/MyDrive/df_fold_1/labels/train_image/02.txt", "r")
lines = file.read().split("\n")
for li in lines[:-1]:
  li = li.split(" ")
  li[0] = int(li[0])
  class_name = names[li[0]]
  li[1] = float(li[1])
  li[2] = float(li[2])
  li[3] = float(li[3])
  li[4] = float(li[4])
  box = [(li[1]-li[3]), (li[2]-li[4]) , (li[1]+li[3]), (li[2]+li[4])]
  #print(box)
  cv2.rectangle(image, (int(box[0]*image_shape[1]), int(box[1]*image_shape[0])), (int(box[2]*image_shape[1]), int(box[3]*image_shape[0])), (0, 255, 0), 2)
  df = df.append({"image_name":image_name, "x1":int(box[0]*image_shape[1]), "y1":int(box[1]*image_shape[0]), "x2":int(box[2]*image_shape[1]), "y2":int(box[3]*image_shape[0]), "class_name":class_name}, ignore_index=True)
cv2_imshow(image)
```
